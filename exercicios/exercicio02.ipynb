```python
import pandas as pd
```

```python
df = pd.read_csv('../dados/titanic.csv')
df
```

```python
#1) Quantas pessoas sobreviveram ao acidente?
df2 = df.loc[df['sobreviveu'] == 'sim']
len(df2.index)
```

```python
#2) Quantas criancas menores de 12 anos sobreviveram ao acidente?
df3 = df.loc[(df['idade'] < 12) & (df['sobreviveu'] == 'sim')]
len(df3.index)
```

```python
#3) Quantas pessoas embarcaram nos portos de Southampton OU Queenstow?
df4 = df.loc[df['embarque'].isin(['Southampton','Queenstow'])]
len(df4.index)
```

```python
#4) Crie um objeto chamado **passageiras** que seleciona somente as passageiras.
def passageiras(row_value):
    if row_value == 'feminino':
        passageira = 1
    else:
        passageira = 0
    return passageira

df['passageira'] = df['sexo'].apply(passageiras)
df.loc[(df['passageira'] == 1)].head()

```

```python
#5) Crie um objeto chamado **criancas_cherbourg** que seleciona as crianças com menos de 12 anos E que embarcaram na cidade de Cherbourg

def criancas_cherbourg(row_value):
    if (row_value[5] < 12) & (row_value[11] == 'Cherboung'):
        crianca_cherbourg = 1
    else:
        crianca_cherbourg = 0
    return crianca_cherbourg

df['criancas_cherbourg'] = df.apply(criancas_cherbourg, axis = 1)
df.head()
```
